```python
# Importing the libraries 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
# Use delimiter=";" as the file uses semicolons to separate values
df = pd.read_csv("student-mat.csv", delimiter=";")
```

```python
df.head()
```

# Data Exploration

```python

# Check for missing values
df.isnull().sum()  
```

```python
# Display column data types
df.dtypes  
```

```python
# Dataset size (number of rows and columns)
df.shape  
```

# Data Cleaning

```python
# Handling missing values - replacing with median if any missing values are present
df_cleaned = df.copy()
df_cleaned.fillna(df_cleaned.median(numeric_only=True), inplace=True)
df_cleaned
```

```python
# Remove duplicate entries
df_cleaned.drop_duplicates(inplace=True)
df_cleaned
```

# Data Analysis Questions

```python
# 1. Calculate the average score in math (G3)
# Use the .mean() method to find the average of the 'G3' column
average_g3 = df['G3'].mean()
print(f"Average score in math (G3): {average_g3:.2f}")
```

```python
# 2. Count how many students scored above 15 in their final grade (G3)
# Filter rows where 'G3' > 15 and count the resulting rows using .shape[0]
students_above_15 = df[df['G3'] > 15].shape[0]
print(f"Number of students scoring above 15 in G3: {students_above_15}")
```

```python
# 3. Check if there is a correlation between study time (studytime) and the final grade (G3)
# Use the .corr() method to calculate the correlation coefficient
correlation = df['studytime'].corr(df['G3'])
print(f"Correlation between study time and G3: {correlation:.2f}")
```

```python
# Scatter plot for visualization
# Create a scatter plot to visualize the relationship between 'studytime' and 'G3'
plt.figure(figsize=(8, 6))  # Set the plot size
sns.scatterplot(x='studytime', y='G3', data=df)  # Plot study time vs. final grade
plt.title('Correlation Between Study Time and Final Grade (G3)')  # Add a title
plt.xlabel('Study Time')  # Label for x-axis
plt.ylabel('Final Grade (G3)')  # Label for y-axis
plt.show()  # Display the plot
```

```python
# 4. Compare the average final grade (G3) by gender
# Group data by the 'sex' column and calculate the mean of the 'G3' column for each group
average_g3_gender = df.groupby('sex')['G3'].mean()
print(f"Average G3 by gender:\n{average_g3_gender}")
```

```python
# Bar plot for visualization
# Create a bar plot to compare the average final grade (G3) by gender
plt.figure(figsize=(8, 6))  # Set the plot size
sns.barplot(x=average_g3_gender.index, y=average_g3_gender.values, palette='viridis')  # Plot gender vs. average grade
plt.title('Average Final Grade (G3) by Gender')  # Add a title
plt.xlabel('Gender')  # Label for x-axis
plt.ylabel('Average Final Grade (G3)')  # Label for y-axis
plt.show()  # Display the plot
```
